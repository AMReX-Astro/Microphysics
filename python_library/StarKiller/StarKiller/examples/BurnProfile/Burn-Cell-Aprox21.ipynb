```python
from StarKiller.initialization import starkiller_initialize
from StarKiller.interfaces import BurnType
from StarKiller.integration import Integrator
from StarKiller.network import Network
import numpy as np
```

```python
probin_file = "probin_aprox21"
```

```python
starkiller_initialize(probin_file)
```

```python
integrator = Integrator()
network = Network()
```

```python
state_in = BurnType()
```

```python
state_in.state.rho = 1.0e9
state_in.state.t = 1.0e9
state_in.state.xn = np.zeros(21)
state_in.state.xn[network.species_map["he4"]] = 1.0
```

```python
print(state_in.ydot)
print(state_in.jac)
print(state_in.state)
```

```python
dt = 1.0e-9
```

```python
out_state = integrator.integrate(state_in, dt)
network.rhs(out_state)
network.jacobian(out_state)
```

```python
print(out_state.ydot)
print(out_state.jac)
print(out_state.state)
```
