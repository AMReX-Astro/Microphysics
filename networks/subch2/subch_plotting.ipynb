```python
import pynucastro as pyna
```

```python
import subch2
lib = subch2.get_library()
```

```python
rc = pyna.RateCollection(libraries=[lib])
```

```python
rc.plot(rotated=True, hide_xalpha=True,
        size=(1500, 450),
        node_size=500, node_font_size=11, node_color="#337dff", node_shape="s",
        Z_range=(1,29))
```

```python
comp = pyna.Composition(rc.get_nuclei())
comp.set_all(0.1)
comp.normalize()
```

```python
rc.plot(rho=1.e6, T=5.e8, comp=comp,
        hide_xalpha=True,
        size=(1200, 1200),
        node_size=500, node_font_size=11, node_color="#337dff", node_shape="s",
        N_range=(5,17), Z_range=(5,17), ydot_cutoff_value=1.e-30)
```
