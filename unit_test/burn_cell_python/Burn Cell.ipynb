```python
import BurnUtils
import numpy as np
```

```python
probin_file = "probin_burn_subch"

time = 10.0
number_of_steps = 1000
```

```python
driver = BurnUtils.BurnerDriver(probin_file)
```

```python
driver.list_species()
```

```python
driver.set_initial_density(1.0e6)

driver.set_initial_temperature(3.0e8)

driver.set_initial_species("he4", 0.83)
driver.set_initial_species("c12", 0.08)
driver.set_initial_species("n14", 0.01)
driver.set_initial_species("o16", 0.08)
```

```python
driver.burn(time, number_of_steps)
```

```python
driver.save("subch")
```
