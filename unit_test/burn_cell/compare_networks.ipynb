```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plot_threshold = 2e-3
```

```python
full_network = "state_over_time.subch2.txt"
approx_network = "state_over_time.subch.txt"
```

```python
full = np.genfromtxt(full_network, names=True)
approx = np.genfromtxt(approx_network, names=True)
```

```python
full_species = list(full.dtype.names)
full_species.remove("Time")
full_species.remove("Temperature")
```

```python
fig, ax = plt.subplots()

cindex = 0

for s in full_species:
    if full[s][1:].max() > plot_threshold:
        lw = 1
        if s in approx.dtype.names:
            lw = 3
        ax.plot(full["Time"], full[s], label=s.capitalize(), color=f"C{cindex}", linewidth=lw)
        if s in approx.dtype.names:
            ax.plot(approx["Time"], approx[s], linestyle=":", color=f"C{cindex}", linewidth=lw)
        cindex += 1
            

ax.set_xscale("log")
ax.set_yscale("log")

ax.set_xlabel("time (s)")
ax.set_ylabel("X")

ax.set_ylim(plot_threshold/1000, 1.0)
ax.legend()

fig.set_size_inches((10, 8))
```
